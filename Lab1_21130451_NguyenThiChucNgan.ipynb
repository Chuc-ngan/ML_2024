```python
class Post:
  def __init__(self, content, subject, likes):
    self.__content = content
    self.__subject = subject
    self.__likes = likes

#get_likes(self)
  def get_likes(self):
    return self.__likes

#get_content(self)
  def get_content(self):
    return self.__content

#get_subject(self)
  def get_subject(self):
    return self.__subject


  def __repr__(self):
    return f"{self.__content},{self.__subject},{self.__likes}"


class Account:
  def __init__(self, name, email, country, friends, posts):
    self.__name = name
    self.__email = email
    self.__country = country
    self.__friends = list()
    self.__posts = list()

  def __repr__(self):
    friends_str = ','.join([friend.get_name() for friend in self.__friends])
    posts_str = ','.join([post.get_content() for post in self.__posts])
    return f"{self.__name},{self.__email},{self.__country},{friends_str},{posts_str}"


#get_name(self)
  def get_name(self):
    return self.__name

#get_email(self)
  def get_email(self):
    return self.__email

#get_country(self)
  def get_country(self):
    return self.__country

#get_friends(self)
  def get_friends(self):
    return self.__friends

#get_posts(self)
  def get_posts(self):
    return self.__posts

  def add_friend(self, friend):
    self.__friends.append(friend)

  def add_post(self, post):
    self.__posts.append(post)

#1) getMaxLikePostByFriend()to find the account having the post with the highest likes from the list of friends.
# tìm tài khoản có bài đăng có lượt thích cao nhất trong danh sách bạn bè.
  def getMaxLikePost(self):
    max_like = 0
    for p in self.get_posts():
      max_like = max(max_like, p.get_likes())
    return max_like

  def getMaxLikePostByFriend(self):
    max_like = 0
    for a in self.__friends:
      max_like = max(max_like, a.getMaxLikePost())
    return max_like

class AccountManager:
  def __init__(self, accounts):
    self.__accounts = accounts

  def add_account(self, accounts):
    self.__accounts.append(accounts)

#2) groupAccountsByPostLike()to group accounts by the number of posts. Where key=false for accounts having less than 10 posts, otherwise key=true (using dictionary).
  def groupAccountsByPostLike(self):
    dis = {False: [], True: []}
    for a in self.__accounts:
      if len(a.get_posts()) >= 10:
        dis[True].append(a)
      else:
        dis[False].append(a)

    return dis

#3) filterAccounts(String country) to get all accounts from a given country. The results are sorted by the number of posts (descending order), and then by name (ascending order).
  def filterAccounts(self, country):
    filter = [a for a in self.__accounts if a.get_country() == country]

    filter.sort(key=lambda a: (-len(a.get_posts()), a.get_name()))
    return filter

# Tạo các đối tượng Post để kiểm tra
post1 = Post("Nội dung bài viết 1", "Chủ đề 1", 50)
post2 = Post("Nội dung bài viết 2", "Chủ đề 2", 20)
post3 = Post("Nội dung bài viết 3", "Chủ đề 3", 30)
post4 = Post("Nội dung bài viết 4", "Chủ đề 4", 15)
post5 = Post("Nội dung bài viết 5", "Chủ đề 5", 45)
post6 = Post("Nội dung bài viết 6", "Chủ đề 6", 25)
post7 = Post("Nội dung bài viết 7", "Chủ đề 7", 35)
post8 = Post("Nội dung bài viết 8", "Chủ đề 8", 10)
post9 = Post("Nội dung bài viết 9", "Chủ đề 9", 55)
post10 = Post("Nội dung bài viết 10", "Chủ đề 10", 40)
post11 = Post("Nội dung bài viết 11", "Chủ đề 11", 18)

# Tạo một số tài khoản
account1 = Account("John", "john@example.com", "USA",  [], [])
account2 = Account("Alice", "alice@example.com", "Canada", [], [])
account3 = Account("Bob", "bob@example.com", "UK", [], [])
account4 = Account("Eve", "eve@example.com", "Australia", [], [])
account5 = Account("Michael", "michael@example.com", "USA", [], [])
account6 = Account("Sophia", "sophia@example.com", "Canada", [], [])
account7 = Account("William", "william@example.com", "Canada", [], [])
account8 = Account("Emma", "emma@example.com", "Canada", [], [])


# Tạo một instance của AccountManager và thêm các tài khoản vào
manager = AccountManager([])

manager.add_account(account1)
manager.add_account(account2)
manager.add_account(account3)
manager.add_account(account4)
manager.add_account(account5)
manager.add_account(account6)
manager.add_account(account7)
manager.add_account(account8)
#print("Post Content: ", post1.get_content())
#print("Post Subject: ",post1.get_subject())
#print("Post Like: ",post3.get_likes())

account1.add_friend(account2)
account1.add_post(post1)
account1.add_friend(account3)
account1.add_post(post3)

account2.add_friend(account1)
account2.add_friend(account3)
account2.add_post(post2)
account2.add_post(post3)
account2.add_post(post4)
account2.add_post(post5)
account2.add_post(post1)
account2.add_post(post6)
account2.add_post(post7)
account2.add_post(post8)
account2.add_post(post10)
account2.add_post(post9)
account2.add_post(post4)
account2.add_post(post11)

account8.add_post(post7)
account8.add_post(post8)
account8.add_post(post10)
account8.add_post(post9)
account8.add_post(post4)
account8.add_post(post11)
#task 1
print("getMaxLikePost: ", account2.getMaxLikePost())
print("getMaxLikePostByFriend:",account2.getMaxLikePostByFriend())

#task2
result = manager.groupAccountsByPostLike()
print("Các tài khoản có ít hơn 10 bài đăng:", result[False])
print("Các tài khoản có ít nhất 10 bài đăng:", result[True])

#task3
filtered_accounts = manager.filterAccounts("Canada")
print("Danh sách các tài khoản từ Canada:")
for account in filtered_accounts:
    print(account.get_name(), "-", account.get_posts())
```
